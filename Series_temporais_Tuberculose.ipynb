```
import pandas as pd

url = "https://raw.githubusercontent.com/alura-cursos/serie_temporal_bootcamp/main/tuberculose_ano_mes_diagnostico.csv"

dados = pd.read_csv(url, encoding = "ISO-8859-1", sep = ";", skiprows=3, skipfooter=16, engine="python")
dados.head()
```

```
dados.tail()
```

```
dados
```

```
dados.drop("Total",axis=1,inplace=True)
dados.drop(19,inplace=True)
```

```
dados
```

# Desafio 01: transformar os dados em uma série temporal Ano-Mês

```
dados_serie_temporal = dados.melt(id_vars="Ano Diagnóstico",var_name="Mês", value_name="Casos")
dados_serie_temporal["data"] = dados_serie_temporal["Ano Diagnóstico"]+"-"+dados_serie_temporal["Mês"]
dados_serie_temporal
```

# Desafio 02: Converter a data (String) em Datatime

```
dados_serie_temporal.info()
```

```
# Para utilizar a transformação do datetime, precisamos converter do portugues
#para o ingles, os meses.
meses_portugues_ingles = {
    "Jan": "Jan",
    "Fev": "Feb",
    "Mar": "Mar",
    "Abr": "Apr",
    "Mai": "May",
    "Jun": "Jun",
    "Jul": "Jul",
    "Ago": "Aug",
    "Set": "Sep",
    "Out": "Oct",
    "Nov": "Nov",
    "Dez": "Dec"
}

dados_serie_temporal["Mês"] = dados_serie_temporal["Mês"].map(meses_portugues_ingles)
dados_serie_temporal.sample(5)
```

```
dados_serie_temporal["data"] = dados_serie_temporal["Ano Diagnóstico"]+"-"+dados_serie_temporal["Mês"]
dados_serie_temporal
```

```
dados_serie_temporal["Data"] = pd.to_datetime(dados_serie_temporal["data"])
dados_serie_temporal
```

```
dados_serie_temporal.info()
```

```
dados_serie_temporal = dados_serie_temporal[["Data", "Casos"]]
dados_serie_temporal
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
sns.lineplot(x="Data", y="Casos", data=dados_serie_temporal)
plt.title("Série Temporal Casos de Tuberculose")
plt.xticks(rotation = 90)
plt.show()
```
