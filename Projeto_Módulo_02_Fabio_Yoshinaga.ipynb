# Coberturas Vacinais por Ano segundo Unidade da Federação
Período: 2009-2019

```
#Leitura do arquivo tabnet.datasus
coberturas_vacinais_por_ano = pd.read_csv('cobertura_vacinas_por_ano.csv', encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=20, decimal=",")

coberturas_vacinais_por_ano
```

```
#Tratamento dos dados:

#Transformar o index em unidade da federação
coberturas_vacinais = coberturas_vacinais_por_ano.set_index('Unidade da Federação')

#Retirar numeral indicativo da UF
coberturas_vacinais.index = coberturas_vacinais.index.str[3:]

#Retirar coluna Total
coberturas_vacinais = coberturas_vacinais.drop('Total', axis=1)

coberturas_vacinais
```

```
#Gráfico coberturas vacinais
coberturas_vacinais.T.plot(figsize=(15,8))
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))
```

Aparentemente a coberturas vacinais no ano de 2015 tiveram uma elevação considerável, chegando em alguns estados utrapassarem as coberturas em mais de 100%, o que nos levam a investigar se houve algum erro na coleta dos dados.

Um outro ponto relevante é a queda em 2016 nas coberturas vacinais, será que é um reflexo do erro em 2015? E foram corrigidos em 2016?

Investigando nos boletins epidemiológicos do SUS (fonte:https://antigo.saude.gov.br/images/pdf/2018/novembro/21/2018-028.pdf), existem fatos confirmados que revelam um erro de imunização. "O erro de imunização (EI) é um tipo de evento adverso pósvacinação (EAPV) considerado evitável. A sua ocorrência
pode se dar por fatores ligados à produção, no não
cumprimento das boas práticas de fabricação; associados
à rede de frio, que compreende erros durante transporte,
armazenamento, acondicionamento, distribuição, controle
de temperatura, entre outros; ou relacionados ao profissional,
por erro na reconstituição, diluição ou dosagem, indicação e
administração das vacinas, validade fora do prazo e outros."

Portanto é possível que um dos fatores para o ano de 2019 terem diminuido as coberturas vacinais em todos os estados, seja este erro de imunização, ou outras variáveis possíveis de relacionamento.
