```
import pandas as pd
import numpy as np


dados = pd.read_csv("/content/hanseniase_diagnostico.csv", encoding="ISO-8859-1", sep=";",
                    skiprows=3, skipfooter=16, engine="python")
dados
```

```
dados.drop("Total",axis=1,inplace=True)
dados
```

```
dados = dados.loc[27:]
dados
```

```
dados_serie_temporal = dados.melt(id_vars="Ano Diagnóstico",var_name="Mês", value_name="casos")
dados_serie_temporal["data"] = dados_serie_temporal["Ano Diagnóstico"]+"-"+dados_serie_temporal["Mês"]
dados_serie_temporal
```

```
meses_portugues_ingles = {
    "Jan": "Jan",
    "Fev": "Feb",
    "Mar": "Mar",
    "Abr": "Apr",
    "Mai": "May",
    "Jun": "Jun",
    "Jul": "Jul",
    "Ago": "Aug",
    "Set": "Sep",
    "Out": "Oct",
    "Nov": "Nov",
    "Dez": "Dec"
}

dados_serie_temporal["Mês"] = dados_serie_temporal["Mês"].map(meses_portugues_ingles)
dados_serie_temporal.sample(5)
```

```
dados_serie_temporal["data"] = dados_serie_temporal["Ano Diagnóstico"]+"-"+dados_serie_temporal["Mês"]
dados_serie_temporal
```

```
dados_serie_temporal["Data"] = pd.to_datetime(dados_serie_temporal["data"])
dados_serie_temporal
```

```
dados_serie_temporal.info()
```

```
dados_serie_temporal["Casos"] = dados_serie_temporal["casos"].astype("int64")
dados_serie_temporal
```

```
dados_serie_temporal.info()
```

```
dados_serie_temporal = dados_serie_temporal[["Data", "Casos"]]
dados_serie_temporal
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
sns.lineplot(x="Data", y="Casos", data=dados_serie_temporal)
plt.title("Série Temporal Casos de Hanseníase")
plt.xticks(rotation = 90)
plt.show()
```

```
dados_serie_temporal.sort_values(by=["Data"], inplace=True)
dados_serie_temporal.reset_index(drop=True, inplace=True)
dados_serie_temporal
```

```
from statsmodels.tsa.arima_model import ARIMA

dados_treino = dados_serie_temporal.query("Ano != 2014")
dados_teste = dados_serie_temporal.query("Ano == 2014")

modelo_arima = ARIMA(dados_treino["Casos"].values.astype("float32"), order=[12,1,2])
modelo_arima_treinado = modelo_arima.fit()
previsao = modelo_arima_treinado.forecast(12)[0]
```

```
dados_teste["previsao"] = previsao
plt.figure(figsize=(10,6))
sns.lineplot(x="Data", y="Casos", data=dados_treino)
sns.lineplot(x="Data", y="Casos", data=dados_teste)
sns.lineplot(x="Data", y="previsao", data=dados_teste)
```

```
from pmdarima.arima import auto_arima

modelo_autoarima = auto_arima(dados_treino["Casos"].values, error_action="ignore", trace=True, seasonal=True, m = 12)
```

```
previsao_autoarima = modelo_autoarima.predict(12)
dados_teste["previsao_autoarima"] = previsao_autoarima
plt.figure(figsize=(10,6))
#sns.lineplot(x="Data", y="Casos", data=dados_treino)
sns.lineplot(x="Data", y="Casos", data=dados_teste)
sns.lineplot(x="Data", y="previsao_autoarima", data=dados_teste)
sns.lineplot(x="Data", y="previsao", data=dados_teste)
```

```
from sklearn.metrics import mean_squared_error

resultado_autoarima = mean_squared_error(dados_teste["Casos"], dados_teste["previsao_autoarima"])
resultado_arima = mean_squared_error(dados_teste["Casos"], dados_teste["previsao"])
print(f"Autoarima = {resultado_autoarima}")
print(f"Arima = {resultado_arima}")
```

```
from fbprophet import Prophet

modelo_profeta = Prophet(seasonality_mode="multiplicative")
mapa = {"Data": "ds", "Casos": "y"}
dados_treino_profeta = dados_treino[["Data", "Casos"]].rename(mapa,axis=1)
modelo_profeta_treinado = modelo_profeta.fit(dados_treino_profeta)
```

```
futuro = modelo_profeta_treinado.make_future_dataframe(periods=12, freq="M")
futuro
```

```
resultado_profeta = modelo_profeta_treinado.predict(futuro)
resultado_profeta
```

```
modelo_profeta_treinado.plot(resultado_profeta);
```
