"O tecido cerebral recebe todas as informações coletadas no meio ambiente através dos sentidos do corpo humano: visão, tato, paladar, olfato e visão. Isso faz com que o tecido cerebral realize um processo, desencadeando na movimentação de músculo. Logo, o contato do cérebro com o meio externo é dado pela da atividade física, por meio de um movimento qualquer.

O exercício físico traz inúmeros benefícios para quem o pratica, podendo ser realizado de modo profissional ou amador, considerando sempre a intensidade do movimento e respeito aos limites do corpo.

Cada pessoa irá optar pela atividade de acordo com sua finalidade. Pode ser um exercício aeróbico, de força ou equilíbrio, contanto que seja realizado de forma organizada, pois contribui com a performance do praticante.

**Importante!** Especialistas alertam que a melhor forma de obter êxito nesse processo é escolher uma atividade que agrade a pessoa que irá realizar.

Durante a realização da atividade recreativa, o cérebro é oxigenado com sangue de boa qualidade e, neste momento, muitas ações ocorrem no corpo, uma delas é a maior atividade cerebral de uma área chamada de hipocampo.

Isso resulta no melhor desempenho da memória e maior capacidade intelectual, além da liberação de endorfina, hormônio que transmite a sensação de prazer, diminui a ansiedade e ativa a serotonina, um neurotransmissor responsável por combater a depressão. "

# Funções Gerais:

```
#DADO = Dataframe criado na função dataframe_por_questionario
#PESQUISA = dado que deseja calcular do Dataframe criado

def frequencia_e_percentual(dado,pesquisa_PERGUNTA):

  frequencia_PERGUNTA = dado[pesquisa_PERGUNTA].value_counts()
  percentual_PERGUNTA = (dado[pesquisa_PERGUNTA].value_counts(normalize = True) * 100).round(2)


  distribuicao_frequencia_PERGUNTA = pd.DataFrame({'Frequencia': frequencia_PERGUNTA, 'Percentual':percentual_PERGUNTA})
  
  return distribuicao_frequencia_PERGUNTA
```

```
#grafico
def grafico_frequencia(dado,pergunta_y, titulo_label_x, titulo_label_y):
  fig = px.bar(x=dado.index, 
              y=dado[pergunta_y],labels={'x':titulo_label_x,'y':titulo_label_y})

  return fig.show()
```

```
#medidas centrais
def medidas_centrais(dado):
  media = dado.mean()
  desvio_padrao = dado.std()
  moda = mode(dado).mode[0]
  mediana = dado.median()
  return media, desvio_padrao, moda, mediana
```

```
from scipy.stats import normaltest

distribuicao_peso_m = questionario.query("SEXO == 'Masculino'")["PESO"]
distribuicao_peso_f = questionario.query("SEXO == 'Feminino'")["PESO"]

_, p_value_normaltest_m = normaltest(distribuicao_peso_m)
_, p_value_normaltest_f = normaltest(distribuicao_peso_f)
print(p_value_normaltest_m)
print(p_value_normaltest_f)
```

Em relação ao peso a distribuição por sexo (masculino e feminino) não é uma distribuição normal, pois o p_value é menor que nosso nível de significancia de 0.05, para um nivel de confiança de 95%.

```
from scipy.stats import ranksums

_, p_value_ranksums = ranksums(distribuicao_peso_m, distribuicao_peso_f)
p_value_ranksums
```

Pelo Peso não ser uma distribuição normal, usamos um teste não parametrico para verificar se a distribuição entre os pesos por sexo tem alguma relação entre si. Como o p_value é menor que 0.05, verificamos que as amostras para o sexo masculino e feminino são independentes uma da outra.

```
from statsmodels.stats.weightstats import DescrStatsW

descricao_imc = DescrStatsW(questionario["IMC_CALCULADO"])
descricao_imc.tconfint_mean()
```

O intervalo de confiança de 95% para o IMC está na faixa de 20.89 a 21.02 que é considerado um nível normal de IMC.

```
from scipy.stats import normaltest

distribuicao_imc_m = questionario.query("SEXO == 'Masculino'")["IMC_CALCULADO"]
distribuicao_imc_f = questionario.query("SEXO == 'Feminino'")["IMC_CALCULADO"]

_, p_value_normaltest_m = normaltest(distribuicao_imc_m)
_, p_value_normaltest_f = normaltest(distribuicao_imc_f)
print(p_value_normaltest_m)
print(p_value_normaltest_f)
```

O IMC entre os sexo apresentam um p_value inferior a 0.05, o que nos classifica como não sendo uma distribuição normal. Faremos um teste não parametrico para comprovar se há relação entre os sexos.

```
from scipy.stats import ranksums

_, p_value_ranksums = ranksums(distribuicao_imc_m, distribuicao_imc_f)
p_value_ranksums
```

Como o p-value é menor que o nivel de significancia do teste, 0.05, rejeitamos a hipótese que haja relação entre os IMC's por sexo da nossa amostra, ou seja, elas são independentes.
